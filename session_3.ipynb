```python
import pandas as pd 

filepath = r"https://raw.githubusercontent.com/data-to-insight/ERN-sessions/main/data/ChildIdentifiers.csv"

child_identifiers = pd.read_csv(filepath)

type(child_identifiers)

child_identifiers.head()
child_identifiers.tail()

child_identifiers

```

```python
#changing datetime fields

child_identifiers["PersonBirthDate_dt"] = pd.to_datetime(child_identifiers["PersonBirthDate"], format="%Y-%m-%d")
child_identifiers["ExpectedPersonBirthDate_dt"] = pd.to_datetime(child_identifiers["ExpectedPersonBirthDate"], format="%Y-%m-%d")

child_identifiers["PersonDeathDate_dt"] = pd.to_datetime(child_identifiers["PersonDeathDate"], format="%Y-%m-%d", errors="coerce")

child_identifiers
```

```python
#max, min, median, mean, sum

max_change = house_prices["Percentage change (monthly) All property types"].max()
min_change = house_prices["Percentage change (monthly) All property types"].min()

print(max_change)
print(min_change)

print(f"max_change : {max_change}, min change: {min_change}")

median_price = house_prices["Average price All property types"].median()
mean_price = house_prices["Average price All property types"].mean()

print(median_price)
print(mean_price)

condition = house_prices["Average price All property types"] < mean_price

less_than_mean = house_prices[condition]

less_than_mean
```

```python
#calculating age column - age is current date - birth date

collection_end = pd.to_datetime("31/03/2025", format = "%d/%m/%Y")

child_identifiers["time_delta"] = collection_end - child_identifiers["PersonBirthDate_dt"]
child_identifiers["Age"] = child_identifiers["time_delta"] / pd.Timedelta(365.25,"d")

child_identifiers["Age"] = child_identifiers["Age"].astype(int)

max_age = child_identifiers["Age"].max()
min_age = child_identifiers["Age"].min()

median_age = child_identifiers["Age"].median()

mean_age = child_identifiers["Age"].mean()

print(max_age)
print(min_age)
print(median_age)
print(mean_age)


```
